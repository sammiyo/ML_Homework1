```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
import numpy as np
import scipy.stats as stats
import math
%matplotlib inline
mu=1
sigma=math.sqrt(2)
mu1=0
sigma1=1
x1=np.linspace(-5,7)
x = np.linspace(-8,8)
plt.plot(x, stats.norm.pdf(x, mu, sigma), label='mean=1,variance=2')
plt.plot(x, stats.norm.pdf(x, mu1, sigma1), label='mean=0,variance=1')
plt.plot(0.840, 0.28, 'o')
plt.plot(-2.8401, 0.007, 'o')
plt.plot([0.840,0.840],[0.28,0],'r--')
plt.plot([-2.840,-2.840],[0.007,0],'r--')
plt.ylabel('P(x|L)')
plt.xlabel('x')
plt.title('Class conditioned pdf')
plt.legend()
plt.show()
plt1.plot(x1,stats.norm.pdf(x, mu, sigma)/((stats.norm.pdf(x, mu, sigma)+stats.norm.pdf(x, mu1, sigma1))), label='Posterior for L=2')
plt1.plot(x1,stats.norm.pdf(x, mu1, sigma1)/((stats.norm.pdf(x, mu, sigma)+stats.norm.pdf(x, mu1, sigma1))), label='Posterior for L=1')
plt1.legend()
plt1.title('Posterior Probability')
plt1.ylabel('P(L|x)')
plt1.xlabel('x')
plt1.show()
```
